Разбейте эти датафреймы на тренировочные (X_train, y_train) и тестовые (X_test, y_test) с помощью
функции train_test_split так, чтобы размер тестовой выборки
составлял 30% от всех данных, при этом аргумент random_state должен быть равен 42.
Создайте модель линейной регрессии под названием lr с помощью класса LinearRegression из модуля
sklearn.linear_model.
Обучите модель на тренировочных данных (используйте все признаки) и сделайте предсказание на
тестовых.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
data = boston['data']
target = boston['target']
features = boston['feature_names']
X = pd.DataFrame(data, columns=features)
y = pd.DataFrame(target, columns=['price'])
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred.shape
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('creditcard.csv')
df.head()
```

```python
pd.options.display.max_columns = 100
```

```python
X = df.loc[:, 'Time':'Amount']
```

```python
X.head()
```

```python
y = df['Class']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
model.fit(X_train, y_train)
```

```python
model.best_params_
```

```python
model.predict_proba(X_test)
```

```python
y_pred_proba = model.predict_proba(X_test)[:, 1]
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, y_pred_proba) #для тестовых данных
```

```python
roc_auc_score(y_train, model.predict_proba(X_train)[:, 1])#для тренировочных данных
```
